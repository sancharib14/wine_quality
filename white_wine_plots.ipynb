### White wine data

#### The plots showing the range of the parameters and the outliers in the data are given below. 

#### For each graph, the colored boxes show the range within which each parameter lies for the given quality index of wine (on the x-axis). The outliers are given by the scattered points above and below the box plots. The bar plot at the end gives a chart of the count of samples against the quality index.

```python
import matplotlib.pyplot as plt
import pandas
import seaborn as sns
```

```python
sns.set(style="darkgrid")

wq = pandas.read_csv('/Users/sancharibanerjee/Downloads/Saama_Summer_Intern_Position/winequality-white.csv', delimiter=';')

for col in wq.columns:
    wq[col]= wq[col].astype('float')

ax = sns.boxplot(x="quality", y="pH", data=wq)
plt.show()
```
